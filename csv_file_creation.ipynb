# File move from src to dest folder

```python
import os
import csv
import shutil
from os import listdir
from os.path import isfile, join

file = open("check_folders.txt", "r", encoding="utf-8")
lines = file.readlines()
dst = "hasan/"
#cnt = 0
for l in lines:
    folder_name = l[0:2]
    path = "mehadi/"+folder_name
    onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
    #cnt += len(onlyfiles)
    for f in onlyfiles:
        src = path + "/" + f
        shutil.move(src, dst)
file.close()
# print(cnt)
```

# Train,Test,Val CSV generate ... and text file creation

```python
import csv
import os
import re

def get_csv(all_folders, csv_file_name, file_path):
    with open(csv_file_name, "w") as csv_file_0:
        for folder_name in all_folders:
            print("****** File name: ",folder_name)
            f = folder_name[0:2]
            with open('all_tsvs/' + folder_name, encoding="utf8") as csv_file:
                csv_reader = csv.reader(csv_file, delimiter='\t')
                for row in csv_reader:
                    flac_file_name = row[0]
                    text = row[2]
                    ##using regex:
                    text = re.sub('[-\t?;:_%!,"a-zA-Z0-9‘’“”।|]*', '', text)
                    text = text.strip()
                    text = text.replace("'", "")
                    text = text.replace("\"", "")
                    text = text.replace("/", " ")
                    print(text)
                    if os.path.exists("wav/"+f+"/"+flac_file_name+".wav") == True: ## jodi wav file exist kore...
                        #csv_creation
                        csv_file_0.write(file_path+"wav/"+flac_file_name+".wav" + "," + file_path+"txt/"+flac_file_name+".txt\n")
                        ##text file creation:
                        file = open(file_path + "txt/" + flac_file_name + '.txt', "w", encoding="utf8")
                        file.write(text)
                        file.close()

##...................................................................................

##For train files
file = open("train_folders.txt", "r", encoding="utf-8")
lines = file.readlines()
train = []
for l in lines:
    train.append(l[0:6]) # new line shoho ashe..
file.close()


# ##For val files
# file = open("val_folders.txt", "r", encoding="utf-8")
# lines = file.readlines()
# val = []
# for l in lines:
#     val.append(l[0:6]) # new line shoho ashe..
# file.close()

# ##For test files
# file = open("test_folders.txt", "r", encoding="utf-8")
# lines = file.readlines()
# test = []
# for l in lines:
#     test.append(l[0:6]) # new line shoho ashe..
# file.close()

# print(len(train))
# print("-----------------------------------------------")
# print(len(val))
# print("-----------------------------------------------")
# print(len(test))

test = ['b4.tsv']

get_csv(test, "google_test_manifest.csv", "Google_Bangla_Dataset/test/")
# get_csv(val, "google_val_manifest.csv", "Google_Bangla_Dataset/val/")
# get_csv(train, "google_train_manifest.csv", "Google_Bangla_Dataset/train/")
```
